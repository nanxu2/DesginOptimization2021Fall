### Problem 3 (30 Points)

Find a local solution to the problem 

$$
\begin{aligned}
\max_{x_1,x_2,x_3} & \quad  f=x_1x_2+x_2x_3+x_1x_3\\
{\text{subject to }} & \quad h=x_1+x_2+x_3-3=0.
\end{aligned}
$$ 

Use two methods: reduced gradient and Lagrange multipliers.

#### Reduced Gradient Methods
The objective function can be converted to minimization problems as:<br />

$$
min_{x} f = -x_1x_2-x_2x_3-x_1x_3
h=x_1+x_2+x_3-3=0
$$

Let $d=\begin{bmatrix}x_1 \\ x_2 \end{bmatrix}$ and $s = x_3 = 3-x_1-x_2$ <br />
$\frac{\partial f}{\partial d}=\begin{bmatrix}-x_2-x_3 \\ x_1 - x_3 \end{bmatrix}$ <br />
$\frac{\partial f}{\partial s}=-x_2-x_1$ <br />
$\frac{\partial h}{\partial d}=\begin{bmatrix}1 \\ 1 \end{bmatrix}$ <br />
$\frac{\partial h}{\partial s}=1$<br />
$\frac{df}{dd}=\frac{\partial f}{\partial d} - \frac{\partial f}{\partial s}(\frac{\partial h}{\partial s})^{-1}\frac{\partial h}{\partial d}=0$<br />
$\frac{df}{dd}=\begin{bmatrix}-x_2-x_3 \\ -x_1 - x_3 \end{bmatrix}+\begin{bmatrix}x_1 + x_2 \\ x_1 +x_2 \end{bmatrix}=0$<br />
Hence $x_1=x_2=x_3$ while $h=x_1+x_2+x_3-3=0$ <br />
Therefore: $\begin{bmatrix}x_1 \\ x_2 \\ x_3 \end{bmatrix}=\begin{bmatrix}1\\1\\1 \end{bmatrix}$<br />
For verification:<br />
$\frac{d^2f}{dd^2}=\begin{bmatrix}I & \frac{ds^T}{dd}\end{bmatrix}\begin{bmatrix}\frac{\partial^2 f}{\partial d^2} & \frac{\partial^2f}{\partial d \partial s} \\ \frac{\partial^2f}{\partial s \partial d} & \frac{\partial^2f}{\partial s^2}\end{bmatrix}\begin{bmatrix}I \\ \frac{ds}{dd}\end{bmatrix}+\frac{\partial f}{\partial s}\cdot{\frac{d^2 s}{dd^2}}$<br />
$\frac{d^2f}{dd^2}=\begin{bmatrix}1 & -1 & -1\end{bmatrix}\begin{bmatrix}0 & -1 & -1 \\ -1 & 0 & -1 \\ -1 & -1 & 0\end{bmatrix}\begin{bmatrix} 1 \\ -1 \\ -1\end{bmatrix} + 0 = 2 > 0$<br />
And:<br />
$\frac{d^2h}{dd^2}=\begin{bmatrix}I & \frac{ds^T}{dd}\end{bmatrix}\begin{bmatrix}\frac{\partial^2 h}{\partial d^2} & \frac{\partial^2h}{\partial d \partial s} \\ \frac{\partial^2h}{\partial s \partial d} & \frac{\partial^2h}{\partial s^2}\end{bmatrix}\begin{bmatrix}I \\ \frac{ds}{dd}\end{bmatrix}+\frac{\partial h}{\partial s}\cdot{\frac{d^2 s}{dd^2}}=0$<br />

#### Lagrange Multipliers
$L = -(x_1x_2+x_2x_3+x_1x_3)+\lambda(x_1+x_2+x_3-3)$<br />
$\frac{\partial L}{\partial x}=\begin{bmatrix}-x_2-x_3+\lambda \\ -x_1-x_3+\lambda \\ -x_1-x_2+\lambda\end{bmatrix} = 0$<br />
$\frac{\partial L}{\partial \lambda}=x_1+x_2+x_3-3=0$<br />
Since we have $x_3 =3- x_1 -x_2$:<br />
$\frac{\partial L}{\partial x}=\begin{bmatrix}x_1 -3+\lambda \\ x_2 -3 +\lambda \\ -x_1-x_2+\lambda \end{bmatrix} = 0$<br />
Hence: <br />
$\begin{bmatrix} x_1 \\ x_2 \\ \lambda\end{bmatrix}=\begin{bmatrix}1\\1\\2 \end{bmatrix}$<br />
Therefore: $\begin{bmatrix}x_1 \\ x_2 \\ x_3 \end{bmatrix}=\begin{bmatrix}1\\1\\1 \end{bmatrix}$<br />
For second order condition:<br />
$dx^TL_{xx}dx=\begin{bmatrix}dx_1 & dx_2 & dx_3 \end{bmatrix}\begin{bmatrix}0 & -1 & -1 \\ -1 & 0 & -1 \\ -1 & -1 & 0 \end{bmatrix}\begin{bmatrix}dx_1 \\ dx_2 \\ dx_3 \end{bmatrix}$<br />
$dx^TL_{xx}dx=-2dx_1dx_2-2dx_1dx_3-2dx_2dx_3=-2dx_1dx_2-2dx_3(dx_1+dx_2)$<br />
From $\frac{\partial h}{\partial x}dx=0$ we can get $dx_3=-dx_1-dx_2$<br />
Hence:<br />
$dx^TL_{xx}dx=-2dx_1dx_2+2(dx_1+dx_2)^2=dx_1^2+dx_2^2+2dx_1dx_2=(dx_1+dx_2)^2$ <br />
where $(dx_1+dx_2)^2 > 0$ for $\forall dx \neq 0$.<br />
Therefore:<br />
$dx^TL_{xx}dx > 0$ for $\forall dx \neq 0$


### Problem 4 (20 Points)

Use reduced gradient to	find the value(s) of the parameter $b$ for which the point $x_1=1$, $x_2=2$ is the solution to the problem 

$$
\begin{aligned}
\max_{x_1,x_2} & \quad  f=2x_{1} + bx_2\\
{\text{subject to }} & \quad g_1 = x_{1}^{2}+ x_{2}^{2}-5\leq 0\\
& \quad g_2= x_1- x_2-2\leq 0.
\end{aligned}
$$
#### Answer
Since $x_1=1$ and $x_2=2$ is the solution of the objective function , let $d=x_1$ and $s=x_2$.

Therefore, we have:<br />
$\frac{df}{dd}=\frac{\partial f}{\partial d} - \frac{\partial f}{\partial s}(\frac{\partial h}{\partial s})^{-1}\frac{\partial h}{\partial d}=0$

where $\frac{\partial f}{\partial d}=2$, $\frac{\partial f}{\partial s}=b$, $(\frac{\partial h}{\partial s})^{-1}=\frac{1}{2x_2}$ and $\frac{\partial h}{\partial d}=2x_1$.

$\frac{df}{dd}=2-b\cdot{\frac{1}{2x_2}}\cdot{2x_1}=0$

$\frac{df}{dd}=2-b\cdot{\frac{1}{4}}\cdot{2}=0$

Hence $b = 4$

### Problem 5 (30 Points)

Find the solution for 

$$
\begin{aligned}
\min_{x_1,x_2,x_3} & \quad  f=x_{1}^{2}+x_{2}^{2}+x_{3}^{2}\\

& \quad h_2 = x_1+x_2-x_3= 0,
\end{aligned}
$$ 

by implementing the generalized reduced gradient algorithm.

```python
from matplotlib import pyplot as plt
import numpy as np

def objfunc(x):
    return np.power(x[0], 2) + np.power(x[1], 2) + np.power(x[2], 2)

def partial_f_partial_d(x):
    return 2 * x[0]

def partial_f_partial_s(x):
    return np.array([2 * x[1], 2 * x[2]])

def partial_h_partial_s(x):
    return np.array([[2/5 * x[1], 2/25 * x[2]], [1, -1]])

def partial_h_partial_d(x):
    return np.array([[1/2 * x[0]],[1]])

def dfdd(x):
    return partial_f_partial_d(x) - np.matmul(np.matmul(partial_f_partial_s(x), np.linalg.inv(partial_h_partial_s(x))), partial_h_partial_d(x))

def x_2_ds(x, a, dfdd_v):

    d = (x[0]-a * dfdd_v)
    s = x[1:3] + a * np.transpose(np.matmul(np.matmul(np.linalg.inv(partial_h_partial_s(x)), partial_h_partial_d(x)), np.transpose([dfdd(x)])))[0]

    return np.append(d, s)

def line_search(dfdd_, x):

    a = 1
    b = 0.5
    t = 0.2

    while objfunc(x_2_ds(x, a, dfdd_)) > (objfunc(x) - a * t * dfdd_ **2):

        a = b * a

    return a

def main(x):

    while np.linalg.norm(np.array([[x[0]**2/4 + x[1]**2/5 + x[2]**2/25 -1 ], [x[0]+x[1]-x[2]]])) > e:

        phps = partial_h_partial_s(x)
        s = np.transpose(np.transpose([x[1:3]]) - np.matmul(np.linalg.inv(phps), np.array([[ x[0]**2/4 + x[1]**2/5 + x[2]**2/25 -1], [x[0]+x[1]-x[2]]])))
        x = np.append(x[0:1], np.transpose(s[0]))

    return x

e = 0.001
x1 = 0
x3 = 1/12 * ( (600-170*(x1**2))**(1/2) + 10*x1)
x2 = x3 - x1

x0 = np.array([x1, x2, x3])
x_ = [x0]
err_ = []
err = 100

while err > e:

    x = x_[-1]
    dfdd_ = dfdd(x)
    err = np.linalg.norm(dfdd_)
    err_.append(err)
    a = line_search(dfdd_, x)
    dk = x[0]- a*dfdd_
    sk = x[1:3] + a* np.transpose(np.matmul(np.matmul(np.linalg.inv(partial_h_partial_s(x)), partial_h_partial_d(x)), np.transpose(dfdd_)))
    xk = np.append(dk, sk)
    x = main(xk)
    x_.append(x)

print('Local solution exists at ' +str(x_[-1]))
plt.plot(err_)
plt.title('Convergence of Error in Semi-log Scaling')
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.yscale('log')
```
